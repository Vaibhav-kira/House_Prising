```python
# in the case we generally use imputation insted of deletion to deal with missing values
# we will treate indepeendent numeric variables....
# we can use mean or median for continuos variable
# for object variables we can use mode
# Raw_Housing_Prices2.csv is the final sheet from target variable treatment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Raw_Housing_Data = pd.read_csv('/content/drive/MyDrive/Notebooks/Module 4 data exploration /2. Data exploration numeric variable/Raw_Housing_Prices2.csv')
```

```python
Raw_Housing_Data.info()
```

```python
# we first create a list of all the integer type data types
numeric_columns = ['No of Bathrooms','Flat Area (in Sqft)','Lot Area (in Sqft)','No of Floors',
                   'Area of the House from Basement (in Sqft)','Latitude','Longitude',
                   'Living Area after Renovation (in Sqft)']
```

```python
from sklearn.impute import SimpleImputer
# imports a class from sklearn.impute i.e. SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy = 'median')
# created an instance imputer here missing_values represent how the missing values are stated in our dataset
# strategy = how we want to replace them
Raw_Housing_Data[numeric_columns] = imputer.fit_transform(Raw_Housing_Data[numeric_columns])
# now we call the function fit_tranform
# fit_transform has two step function 
# fit --> It finds the median of all the columns that are passes and stores them ....
# transform --> this parts does the actual action of finding the missing values
# and imputing them with the median  
```

```python
Raw_Housing_Data.info()
```

```python
imputer2 = SimpleImputer( missing_values = np.nan , strategy = 'most_frequent')
# Raw_Housing_Data['Zipcode'] = imputer2.fit_transform(Raw_Housing_Data['Zipcode'])
# will get an error that Expected 2d array got 1D 
# we can reshape our array to get around this
Raw_Housing_Data['Zipcode'] = imputer2.fit_transform(Raw_Housing_Data['Zipcode'].values.reshape(-1,1))
```
