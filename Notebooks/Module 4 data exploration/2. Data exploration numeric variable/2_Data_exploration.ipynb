```python
# No of times visited is a variable which would be best descibed in the form of numeric
Raw_Housing_Data['No of Times Visited'].unique()
```

```python
# recode then into numeric using pandas . map function
mapping = {'None' : '1',
           'Thrice' : '3',
           'Twice' : '2',
           'Once' : '1' ,
           'Four' : '4'}
Raw_Housing_Data['No of Times Visited'] = Raw_Housing_Data['No of Times Visited'].map(mapping)
```

```python
Raw_Housing_Data['No of Times Visited'].unique()
```

```python
Raw_Housing_Data['No of Times Visited'] = Raw_Housing_Data['No of Times Visited'].astype(int)
Raw_Housing_Data['No of Times Visited'].unique()
```

```python
# deriving new variable from existing variable
# we want to derive two new variabes from the type Renovated year
# two variable ---> Ever Renovated , Year Since renovation
# Ever Renovated --> Yes , No 
```

```python
# when value of renovated year ==0 No is assigned else yes 
Raw_Housing_Data['Ever Renovate'] = np.where(Raw_Housing_Data['Renovated Year'] == 0 , 'No' , 'Yes')
Raw_Housing_Data.head(5)
```

```python
# For Year Since Renovation 
# Year of Sale - Year of renovation
# if Year of renovation = 0 
# for never renovated Year since  = 0
# numerical variable
```

```python
# Extracting the year of the house sell 
Raw_Housing_Data['Purchase year'] = pd.DatetimeIndex(Raw_Housing_Data['Date House was Sold']).year
```

```python
Raw_Housing_Data['Years Since Renovation'] = np.where(Raw_Housing_Data['Ever Renovate'] == 'Yes',
                                                      abs(Raw_Housing_Data['Purchase year'] - Raw_Housing_Data['Renovated Year']),
                                                      0 )
```

```python
# To keep the data clean we can drop some columns 
# like Renovated year 
# Purchase year 
# and Date House was sold
Raw_Housing_Data.drop(columns=['Date House was Sold' , 'Purchase year' , 'Renovated Year' ],inplace= True)
Raw_Housing_Data.head(10)
```
