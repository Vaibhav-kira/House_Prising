```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
Data = pd.read_csv("/content/drive/MyDrive/Notebooks/Module 3/libraries/Pandas specific/1. Regression - Module - (Housing Prices).csv")
```

```python
Data
```

```python
# initializing a new colum
Data['Condion_Sale'] = 0

# Calculating mean on the basis of the condition of the house
for i in Data['Condition of the House'].unique():
  Data['Condion_Sale'][Data['Condition of the House'] == str(i)] = Data['Sale Price'][Data['Condition of the House'] == str(i)].mean()

# plotting the mean sales based on the condition of the house
plt.figure(dpi=100)
plt.bar(Data['Condition of the House'].unique(),Data['Condion_Sale'].unique())
plt.xlabel("Condition of the house")
plt.ylabel("Mean Sale Price of House")
plt.show()
```

```python
# with respect to zipcode and condition of the house PRICE
Data['Zipcode_condition_sale'] = 0
Z = 'Condition of the House'
Y = 'Zipcode'
X = 'Zipcode_condition_sale'
P = 'Sale Price'
for i in Data['Condition of the House'].unique():
  for j in Data['Zipcode'].unique():
    Data['Zipcode_condition_sale'][(Data['Condition of the House'] == str(i)) & (Data['Zipcode'] == j)] = Data[P][(Data[Z] == str(i)) & (Data[Y] == j)].mean()
len(Data[X].unique())
```

```python
# above using pandas
Zip_Condition_Sale = Data.groupby(['Condition of the House','Zipcode'])['Sale Price'].mean()
Zip_Condition_Sale
```

```python
Zip_Condition_Sale2 = pd.pivot_table(Data,index = ['Condition of the House','Zipcode'],values=['Sale Price'],aggfunc = np.mean)
Zip_Condition_Sale2
```

```python
# better way
Zip_Condition_Sale3 = pd.pivot_table(Data,index = ['Zipcode'],columns=['Condition of the House'],values=['Sale Price'],aggfunc = np.mean)
Zip_Condition_Sale3
```

```python
# Adding plot length
Data['Plot Length'] = Data['Flat Area (in Sqft)']**0.5
Data['Plot Length'].head()
```

```python
# as the mean of good and fair are comparitive the can be replaced with common value
Data['Condition of the House'] = Data['Condition of the House'].map(
    {
        'Good' : '1',
        'Fair' : '1',
        'Bad'  : '0',
        'Okay' : '1',
        'Excellent' : '3',     
    }
)
Data['Condition of the House'].unique()
```

```python
# date House was sold extract year to another column
year = []
for i in range(len(Data['Date House was Sold'])):
  k = Data['Date House was Sold'][i].split()[-1]
  year.append(k)
Data['Year_Sold'] = year
Data['Year_Sold'].head()
```

```python
# above using pandas . map
def yearf(v):
  return v.split()[-1]
Data['Year_Sold'] = Data['Date House was Sold'].map(yearf)
Data['Year_Sold'].head()
```

```python
# Luxuary home using basic python script
Data['Luxury'] = 0
for i in range(len(Data)):
  count = 0
  if Data['Waterfront View'][i] == 'Yes':
    count = count + 1
  if Data['Condition of the House'][i] in ['Good','Excellent']:
    count = count + 1
  if Data['Overall Grade'][i] >= 8:
    count = count + 1
  if count >= 2:
    Data['Luxury'][i] = 'Yes'
  else:
    Data['Luxury'][i] = 'No'
Data['Luxury'].unique()
```

```python
# using pandas.apply MMMUUUUUCCCCHHHH FASTER
def Luxury(rows):
  count = 0
  if rows[0] == 'Yes':
    count += 1
  if rows[1] in ['Good','Excellent']:
    count += 1
  if rows[2] >= 8:
    count += 1
  if count>=2:
    return 'Yes'
  else:
    return 'No'

Data['Luxury'] = Data[['Waterfront View','Condition of the House','Overall Grade']].apply(Luxury,axis = 1)
Data['Luxury'].unique()
```
